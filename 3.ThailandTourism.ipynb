# About the data
This dataset contains statistics on domestic tourism in Thailand from Jan 2019 to Feb 2023, broken down by province. The dataset includes information on the number of tourists, the occupancy rate, and the profits generated by tourism in each province, as well as the nationality of the tourists (Thai vs. foreign).

The purpose of this notebook is to perform exploratory data analysis finding insights in thailand tourism and answering some questions such as , which region is the most popular among foreign tourists.

# Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import datetime as dt
sns.set_style('whitegrid')
%matplotlib inline
pd.set_option('display.max_rows', 50)
import warnings
warnings.filterwarnings("ignore")
```

```python
url = 'https://github.com/Kodchalerk/EDA-ThailandTourism/blob/main/thailand_domestic_tourism_2019_2023.csv?raw=True'
df = pd.read_csv(url)
df
```

# Data Cleaning and processing

# Data Exploratory

## Let's find out all about occupancy rate

```python
df_occupancy = df[df['variable'] == 'occupancy_rate']
df_occupancy
```

```python
# Which region has the higest average occupancy rate
df_occupancy1 = df_occupancy.groupby('region_eng').agg(average_occupancy_rate = ('value','mean')).sort_values('average_occupancy_rate', ascending = False)
```

```python
sns.barplot(data = df_occupancy1 , x = df_occupancy1.index , y = 'average_occupancy_rate' , palette = sns.color_palette("hls", 8))
plt.title('Average occupancy rate in each region between 2019 and 2023')
plt.xlabel('Region')
plt.ylabel('Average Occupancy Rate');
plt.savefig('TH1.png')
```

- Now we know that in the **northern region** has the highest occupancy rate between 2019 and 2023.
- Let's further investigate which province in the north has highest average occupancy rate.

```python
df_top5_occupancy = df_occupancy[df_occupancy['region_eng'] == 'north'].groupby('province_eng').agg(average_occupancy_rate = ('value','mean'))
df_top5_occupancy = df_top5_occupancy.sort_values('average_occupancy_rate' , ascending = False).head(5)

sns.barplot(data = df_top5_occupancy , x = df_top5_occupancy.index , y = 'average_occupancy_rate' , palette = sns.color_palette("husl", 8))
plt.title('Top 5 average occupancy rate in Northern Region between 2019 and 2023')
plt.xlabel('Province')
plt.ylabel('Average Occupancy Rate')
plt.savefig('TH2.png')
```

- The figure above indicated that Nan province has the highest average occupancy rate betwee 2019 and 2023 
- Next, we will investigate the province that has higest average occupancy rate.

```python
top_province = df_occupancy.groupby(['province_eng','region_eng'], as_index = False).agg(average_occupancy_rate = ('value','mean'))
# top_province = pd.DataFrame(top_province.sort_values('average_occupancy_rate', ascending = False).head(10))
top_province = pd.DataFrame(top_province.nlargest(10,'average_occupancy_rate'))
top_province = top_province.sort_values('average_occupancy_rate' , ascending = True)
```

```python
sns.set_style('darkgrid')
barplot = plt.barh(data = top_province , y = 'province_eng' , width = 'average_occupancy_rate' )
plt.title('Top 10 average occupancy rate between between 2019 and 2023')
plt.xlabel('Province')
plt.ylabel('Average Occupancy Rate')
plt.bar_label(barplot, labels = top_province.region_eng , label_type = 'center')
####Differiate colour by region
## North East
barplot[0].set_color('aquamarine')
barplot[1].set_color('aquamarine')
barplot[3].set_color('aquamarine')
# North
barplot[9].set_color('salmon')
barplot[7].set_color('salmon')
barplot[5].set_color('salmon')
barplot[2].set_color('salmon')
# Central
barplot[8].set_color('lightgoldenrodyellow')
barplot[6].set_color('lightgoldenrodyellow')
barplot[4].set_color('lightgoldenrodyellow')

plt.savefig('TH3.png')
```

- Surprisingly, there is no province from the southern region in this chart.

## Let's find out the trend of tourism in Thailand between 2019 and 2023

```python
# Extracting Date column
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
```

```python
df
```

```python
# Prepare dataframe for the analysis of trend
df_no_tourist = df[df['variable'] == 'no_tourist_all']
df_no_tourist
```

```python
tourist = df_no_tourist.groupby(['date','region_eng'], as_index = False).agg(number_of_tourist = ('value','sum'))
tourist_central = tourist[tourist['region_eng'] == 'central']
tourist_north = tourist[tourist['region_eng'] == 'north']
tourist_east_northeast = tourist[tourist['region_eng'] == 'east_northeast']
tourist_east = tourist[tourist['region_eng'] == 'east']
tourist_south = tourist[tourist['region_eng'] == 'south']
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x= tourist_central.date , y = tourist_central.number_of_tourist , label = 'Central')
sns.lineplot(x= tourist_north.date , y = tourist_north.number_of_tourist , label = 'North')
sns.lineplot(x= tourist_east_northeast.date , y = tourist_east_northeast.number_of_tourist , label = 'Northeast')
sns.lineplot(x= tourist_east.date , y = tourist_east.number_of_tourist , label = 'East')
sns.lineplot(x= tourist_south.date , y = tourist_south.number_of_tourist , label = 'South')

###
plt.title('Number of tourists between 2019 and 2023')
plt.xlabel('Date')
plt.ylabel('Number of Tourists')
plt.xticks(rotation = 45)
plt.savefig('TH4.png')
```

- It seems that the tourism in each region has been gradually recovering to it's previous state before the COVID outbreak.
- However, the tourism in southern region seems to face diffucties in recovering back to it's glorious period at 2019.

## Let's find out which province and region is the most popular among local and foreign tourists.

```python
df.variable.unique()
```

```python
df_local = df[df['variable'] == 'no_tourist_thai']
df_foreign = df[df['variable'] == 'no_tourist_foreign']
```

```python
df_local
```

```python
local = df_local.groupby('region_eng', as_index = False).agg(average_tourist = ('value','mean'))
local = local.sort_values('average_tourist',ascending = False)
local
```

```python
sns.barplot(data = local , x = 'region_eng' , y = 'average_tourist' , palette = sns.color_palette("husl", 8))
plt.title('Average number of local tourists in each region between 2019 and 2023')
plt.xlabel('Province')
plt.ylabel('Average Number of Tourists')
```

```python
foreign = df_foreign.groupby('region_eng', as_index = False).agg(average_tourist = ('value','mean'))
foreign = foreign.sort_values('average_tourist',ascending = False)
foreign
```

```python
sns.barplot(data = foreign , x = 'region_eng' , y = 'average_tourist' , palette = sns.color_palette("husl", 8))
plt.title('Average number of foreign tourists in each region between 2019 and 2023')
plt.xlabel('Province')
plt.ylabel('Average Number of Tourists')
```

```python
fig , (ax1,ax2) = plt.subplots(nrows = 1 , ncols = 2 , figsize = (10,5) , sharey = 'row')
fig.suptitle('The popularity of region between local and foreign tourist')

#ax‡πÖ
figure1 = ax1.bar(data = local , x = 'region_eng' , height = 'average_tourist')
figure1[0].set_color('aquamarine')
figure1[1].set_color('salmon')
figure1[2].set_color('lime')
figure1[3].set_color('violet')
figure1[4].set_color('azure')
ax1.set_xticklabels(labels = local.region_eng,rotation = 45)
ax1.set_ylabel('Average Number of Tourists')
ax1.set_title('Local Tourists')

#ax2
figure2 = ax2.bar(data = foreign , x = 'region_eng' , height = 'average_tourist')
figure2[0].set_color('aquamarine')
figure2[1].set_color('azure')
figure2[2].set_color('salmon')
figure2[3].set_color('violet')
figure2[4].set_color('lime')
ax2.set_xticklabels(labels = foreign.region_eng,rotation = 45)
ax2.set_title('Foreign Tourists')
##
plt.tight_layout(w_pad = 3)
plt.savefig('TH5.png')
```

- This figure shows the average number of tourists from 2019 to 2023 between local and foreign tourists.
- As we can see from the figures, the most popular region among the local tourists is in the central region similar with the foreign tourists.
- The second most popular region for local tourists is in the eastern region while the foreign tourists are in favor of southern region.
- Surprisingly, the south area is the least popular region among the local tourists.
- In the next section we will investigate further the most attractive provinces among two group of tourists.

```python
# Create dataframe for local tourists' favourite provinces
local_province = df_local.groupby(['province_eng','region_eng'] , as_index = False).mean('value').sort_values('value' , ascending = False).head(5)
local_province
```

```python
# Create dataframe for foreigns tourists' favourite provinces
foreign_province = df_foreign.groupby(['province_eng','region_eng'] , as_index = False).mean('value').sort_values('value' , ascending = False).head(5)
foreign_province
```

```python
fig , (ax1,ax2) = plt.subplots(nrows = 1 , ncols  = 2 , figsize = (15,5) , sharey = True )
fig.suptitle('Top 5 popular provinces among local and foreign tourists')
#ax1 local tourists

figure1 = ax1.bar(data = local_province, x = 'province_eng' , height = 'value')

ax1.set_xticklabels(labels = local_province.province_eng,rotation = 45)
ax1.set_ylabel('Number of Tourists')
ax1.set_title('Local Tourists')
ax1.bar_label(figure1 , labels = local_province.region_eng , padding = 2)

figure1[0].set_color('aquamarine')
figure1[1].set_color('aquamarine')
figure1[2].set_color('salmon')
figure1[3].set_color('aquamarine')
figure1[4].set_color('lime')

#ax2 Foreign tourists

figure2 = ax2.bar(data = foreign_province, x = 'province_eng' , height = 'value')
ax2.set_title('Foreign Tourists')
ax2.set_xticklabels(labels = foreign_province.province_eng,rotation = 45)
ax2.bar_label(figure2 , labels = foreign_province.region_eng );

figure2[0].set_color('aquamarine')
figure2[1].set_color('azure')
figure2[2].set_color('salmon')
figure2[3].set_color('violet')
figure2[4].set_color('azure')

plt.savefig('TH6.png')
    
```

- This figure demonstrated the average number of tourists in the top 5 most popular province from 2019 to 2023.
- It seems that Bangkok, the capital city , is the most popular choice between two groups of tourist.
- For local tourists, there is no province from nothern and southern region present in the top 5 most popular provinces while nort-eastern region seems not to be popular among foreign tourists.

## Let's take a look at the perspective of profit generation

```python
# Prepare dataframe for analysis

df_net_all = df[df['variable'] == 'net_profit_all']
df_net_thai = df[df['variable'] == 'net_profit_thai']
df_net_foreign = df[df['variable'] == 'net_profit_foreign']
```

```python
df_net_all = df_net_all.groupby(['date','region_eng'], as_index = False).sum()
df_net_all
```

```python
sns.lineplot(data = df_net_all , x = 'date' , y = 'value' , hue = 'region_eng')
plt.title('Profit generated by local tourism in each region (Million THB)')
plt.ylabel('Profit')
plt.xlabel('Date')
plt.xticks(rotation = 45);
```

```python
df_net_thai = df_net_thai.groupby(['date','region_eng'], as_index = False).sum()
df_net_thai
```

```python
plt.figure(figsize = (10,5))
sns.lineplot(data = df_net_thai , x = 'date' , y = 'value' , hue = 'region_eng' , palette = sns.color_palette("pastel"))
plt.title('Profit gener|ated by local tourism in each region (Million THB)')
plt.ylabel('Profit')
plt.xlabel('Date')
plt.xticks(rotation = 45);
```

```python
df_net_foreign = df_net_foreign.groupby(['date','region_eng'], as_index = False).sum()
```

```python

plt.figure(figsize = (10,5))
sns.lineplot(data = df_net_foreign , x = 'date' , y = 'value' , hue = 'region_eng' , palette = sns.color_palette("husl", 9))
plt.title('Profit generated by international tourism in each region (Million THB)')
plt.ylabel('Profit')
plt.xlabel('Date')
plt.xticks(rotation = 45);
```

```python
df_net_gross = df_net_all.groupby('region_eng' , as_index = False).sum()
df_net_gross = df_net_gross.sort_values('value' , ascending = False)
```

```python
sns.barplot(data = df_net_gross , x = 'region_eng' , y = 'value')
```

```python
fig = plt.figure(figsize = (20,10))

fig.suptitle('Net profit generated by tourism in Thailand from 2019 to 2023')

#ax2
ax2 = fig.add_subplot(2,2,2)
sns.barplot(data = df_net_gross , x = 'region_eng' , y = 'value' , ax = ax2 , palette = 'viridis')
ax2.set_ylabel('Profit (Million THB)')
ax2.set_title('Net profit generated from tourism in Thailand between 2019 and the beginning of 2023')
ax2.set_xlabel('Region')

#ax1
ax1 = fig.add_subplot(2,2,1)
sns.lineplot(data = df_net_all , x = 'date' , y = 'value' , hue = 'region_eng' , palette = sns.color_palette("hls", 8) ,ax = ax1 )
ax1.set_ylabel('Profit (Million THB)')
ax1.set_title('Net profit')

#ax4

ax4 = fig.add_subplot(2,2,4 , sharey = ax4)
sns.lineplot(data = df_net_thai , x = 'date' , y = 'value' , hue = 'region_eng' , palette = sns.color_palette("hls", 8) , ax = ax4)
ax4.set_ylabel('Profit (Million THB)')
ax4.set_title('Net profit generated by Local Tourism')

#ax3
ax3 = fig.add_subplot(2,2,3 , sharey = ax4)
sns.lineplot(data = df_net_foreign , x = 'date' , y = 'value' , hue = 'region_eng' , palette = sns.color_palette("hls", 8) ,ax = ax3)
ax3.set_ylabel('Profit (Million THB)')
ax3.set_title('Net profit generated by International Tourism')






plt.tight_layout()
plt.savefig('TH7.png')
```

- Overall, provinces in central region generated the highest amount of net profit followed by south , east , north ,and north-east , respectivey.
